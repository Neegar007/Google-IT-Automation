### Solving Problems

* Gather Necessary Information. E.g. reproduction case. i.e a clear
description of how and when the problem appears.
* Finding the root cause. 
* Performing the neccesary remediation. a short term solution to get the 
system running again, and the long-term solution to prevent such occurence 
in the future.
* We also need to document the steps we take in solving the problems.

### Tools For System Troubleshooting

* strace. This trace all the system calls a script or program made and tell
the result of these calls. Good for debugging crashing apps. Systems calls are
all the calls our program made to the kernel. Run the code below to save the 
output to a file and use 'less' to open it
$ strace -0 outputfile.strace ./file_to_trace.py 

### Reproduction Case

* A reproduction case is a way to verify is the problem is present or not.
Another important step is to get to know the root cause of the problem. This is very critical for long-term solution.
* We separate a test environment from production so we don't break things
our users are working with unintentionally.
