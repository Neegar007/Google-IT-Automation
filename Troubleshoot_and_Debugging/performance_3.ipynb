### Caching

* Store a result once → reuse it instead of recomputing

```python
from pymemcache.client import base
```

```python
from pymemcache.client import base

client = base.Client(('localhost', 11211))

# Save to cache
client.set('price_btc', '65000')

# Read from cache
price = client.get('price_btc')

print(price)
```
