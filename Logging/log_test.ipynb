```python
# Making Sense of the log file
# This script will read through the log file and count the number 
# of occurrences of each type of HTTP request method users made to the server.
Report = {}  # Dictionary to hold the report data
pattern = r"(\"[A-Z]+)\s"

with open("../logfiles.log") as f:
        for line in f:
          search = re.search(pattern, line)
          # This line updates the Report dictionary with the count 
          # of each request method. If the method is already in the 
          # dictionary, it increments its count by 1; otherwise, it adds 
          # the method to the dictionary with an initial count of 1.
          Report[search[1].strip('"')] = Report.get(search[1].strip('"'), 0) + 1
print(Report)
```
