```python
import subprocess
# The 'who' command shows who is logged in
```

```python
result = subprocess.run(['host', 'google.com'], capture_output=True)
print(result.returncode)
# The output is in bytes, so we decode it to a string and 
# captured in the stdout attribute
print(result.stdout.decode().split())
```

```python
# stderr will capture any error messages
result = subprocess.run(['rm', 'nonexistent_file'], capture_output=True)
print(result.returncode)
print(result.stderr.decode())
```
