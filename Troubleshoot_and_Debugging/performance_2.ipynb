### Parallelism

* When we are performing IO operations over a network or disk, the CPU remains idle. And sometimes, the execution of our script is like a one way traffic. The script is blocked while waiting for input or output operations. To solve this, we can do operations in parallel. i.e while waiting for the IO operations on disk or network, we can be running operations on the cpu. At the end, everything will come together to give us the same result.
   
*   A very easy way to run operations in parallel is just to split them across different processes, calling your script many times each with a different input set, and just let the operating system handle the concurrency. 
*   Sometimes these processes share data, that's when 'threads' comes in. Threads let us run parallel task inside a process. Since this isn't handled by the OS, we'll need to modify our code to create and handle the threads. For that, we'll need to look into how the programming language we're using implements threading. In Python, we can use the Threading or AsyncIO modules to do this. These modules let us specify which parts of the code we want to run in separate threads or as separate asynchronous event and how we want the result of each to be combined in the end.

#### Example of paralellism
*   We do this for CPU bound operations like data manipulations, scientific 
computations, complex logic calculations.

```python
# parallel_example.py

from multiprocessing import Pool, cpu_count
import time
from datetime import datetime


def square(number):
    # Operations to be executed in parallel by different processes.
    print(f"Processing {number}")
    time.sleep(1)  # simulate heavy work
    return number * number

start = datetime.now().second

if __name__ == "__main__":
    # Items to perform the computation on. Could be files, data from network 
    # or database
    numbers = [1, 2, 3, 4, 5] 

    # Create a pool of worker processes
    with Pool(cpu_count()) as pool:
        results = pool.map(square, numbers)

    print("Results:", results)

close = datetime.now().second
print(f"Execution time: {close - start} seconds")
```

#### Example of Concurrency
*   We use this when performing IO boud operations. Like reading and writing files to the disk or network. threading and asyncio perform this BUT the difference is that threading creates multiple threads to carryout operations on the memory at the same time just like processes on CPU WHILE asyncio is just one thread that performs many operations on the memory. Asyncio achieve this by moving from one task to another while waiting for the former task to complete.
    Asyncio is prefered because in the case of 1000 IO operations, threading will overload the memory with 1000 threads each working at the same time. While asycncio uses one thread to perform the operations simuteneously. This makes it memory efficient. Areas good for asyncio :

    *   Network requests
    *   Database calls
    *   Reading sockets
    *   APIs
    *   Streaming logs
    *   Bots

```python
import asyncio
from datetime import datetime
async def task():
    # Simulate an asynchronous operation, such as a network request or file I/O
    # i.e while waiting for the operation to complete, other tasks 
    # can run concurrently.
    await asyncio.sleep(2) 
    print("Done")

async def main():
    await asyncio.gather(
        task(),
        task(),
        task()
    )
start = datetime.now().second
await main()
close = datetime.now().second
print(f"Execution time: {close - start} seconds")
```

### Common Blocking Functions (Danger List)

* These secretly break asyncio:

| Blocking         | Async Alternative       |
| ---------------- | ----------------------- |
| `time.sleep()`   | `await asyncio.sleep()` |
| `requests.get()` | `aiohttp`               |
| `open().read()`  | `aiofiles`              |
| heavy loops      | multiprocessing         |

Whenever we want to perform concurrent tasks, we use the asyncio alternatives
