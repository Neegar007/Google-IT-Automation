## Exception handling prevent you from wriring multiple If/Else statements

```python
#!/usr/bin/env python3

def character_frequency(filename):
  """Counts the frequency of each character in the given file."""
  # First try to open the file
  try:
    f = open(filename)
  except OSError:
    return None

  # Now process the file
  characters = {}
  for line in f:
    for char in line:
      characters[char] = characters.get(char, 0) + 1
  f.close() 
  return characters

char_freq = character_frequency("../Practice/get_request.log")
print(char_freq)
```

#### Raising Your Own Errors

```python
def validate_user(username, minlen):
  if minlen < 1:
    # Must raise the correct exception. In this case, ValueError is appropriate
    raise ValueError("minlen must be at least 1")

  if len(username) < minlen:
    return False
  if not username.isalnum():
    return False
  return True
```

```python
validate_user("", -1)
```

```python
validate_user("", 1)
validate_user("myuser", 1)
```

```python
validate_user(88, 1)
```

```python
validate_user([], 1)
```

```python
validate_user(["name"], 1)
```

```python
def validate_user(username, minlen):
  # Behavioural assertions. Meaning we are checking for correct behaviour
  assert type(username) == str, "username must be a string"
  if minlen < 1:
    raise ValueError("minlen must be at least 1")

  if len(username) < minlen:
    return False
  if not username.isalnum():
    return False
  return True
```

```python
validate_user([3], 1)
```

```python
my_list = [27, 5, 9, 6, 8]
def RemoveValue(myVal):
    if myVal not in my_list:
        raise ValueError(myVal + " no longer in the list")
    else:
        my_list.remove(myVal)
    return my_list

print(RemoveValue(27))
```

```python
import random

participants = ['Jack','Jill','Larry','Tom']

def Guess(participants):
    my_participant_dict = {}
    for participant in participants:
        my_participant_dict[participant] = random.randint(1, 9)
    if my_participant_dict['Larry'] == 9:
        return True
    else:
        return False
    
print(Guess(participants))
```

```python
# Revised Guess() function
def Guess(participants):
    my_participant_dict = {}
    try:
        for participant in participants:
            my_participant_dict[participant] = random.randint(1, 9)
        if my_participant_dict['Larry'] == 9:
            return True
        else:
            return False
    except KeyError: # Catcth the correct error to replace System error message
        return None
```

```python
participants = ['Cathy','Fred','Jack','Tom']
print(Guess(participants))
```
