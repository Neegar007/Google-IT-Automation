### Puppet

*   Puppet is used to write configuration files to configure servers. It's a client-server achitecture where the server will send machine info to the server and the server will send config files to the client to implement.

### Classes, Resources, Manifest
  
*   Resources are the basic unit for modeling the configuration that we want to manage. Like service, package, file, exec, cron, group, notify, user.Custom recources can also be written.

*   Classes are group of related resources to be managed. 
*   Resources and classes are written in manifest files - *.pp
*   Modules are directory-tree containing manifests and metafiles. All packages organised in a way that your system understands.

```python
class sudo {
 
  package { 'sudo':
    ensure => present,
  }
}

#   This means the package 'sudo' will be installed on every server where the rules are applied. Be it 1000 servers.
```

####  Define a file resource and ensure that it's a directory

```python
class sysctl {
  # Make sure the directory exists, some distros don't have it
  file { '/etc/sysctl.d':
    ensure => directory,
  }
}

#
```

*   In this code block, we are configuring the contents of /etc/timezone.This will be a file, and the contents of the file will be set to the UTC timezone. We also set the replace attribute to true which means even if the contents of the file already exist, they will  be replaced

```python
class timezone {

      file { '/etc/timezone':
        ensure  => file,
        content => "UTC\n",
        replace => true,
      }

}
```
