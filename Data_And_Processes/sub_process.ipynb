### Running System Program

```python
import subprocess
subprocess.run(["date"])
```

```python
"""
The script is the parent process that creates a child process 
to execute commands. After the child process completes, it returns 
a return code to the parent process indicating success (0) or failure (non-zero).

"""
```

```python
subprocess.run(["date"])
subprocess.run(["sleep", "5"])
```

```python
"""
We added a sleep command to pause before returning control to 
the parent process. The parent process is blocked until the child
process completes its execution. After the sleep duration, the child process
completes and returns control to the parent process to continue its execution.
"""
```

```python
subprocess.run(["date"])
subprocess.run(["sleep", "2"])
# Linux command for listing files in a directory : ls
result = subprocess.run(["ls", "../RegExp"])
print("Return Code:", result.returncode)
print(type(result))
```

### Advanced Subprocesses

```python
import os
import subprocess
```

```python
# Return a dictionary representing the user's environment variables and copied it
my_env = os.environ.copy()
# Select the PATH variable and prepend a new directory to it
# Telling the subprocess to look in this directory for executables
my_env['PATH'] = os.pathsep.join(['/home/devops/automation/Data_And_Processes', my_env['PATH']])
# Run the command with the modified environment
result = subprocess.run(['my_script.sh'], env=my_env)
```

```python
result = subprocess.run(['host', 'neolife.com'], capture_output=True, text=True)
print(result.stdout.split())
```
