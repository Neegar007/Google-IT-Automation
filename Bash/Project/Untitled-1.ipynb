```python
import re
from collections import defaultdict
```

```python
log = "Jan 31 00:21:30 ubuntu.local ticky: ERROR The ticket was modified while updating (breee)"
msg = {}
per_user = defaultdict(lambda: {"ERROR": 0, "INFO": 0})
pattern = re.compile(r"(ERROR|INFO)\s.*\((\w+)")
match = pattern.search(log)
msg_level, user = match.groups()
per_user[user][msg_level] = per_user.get(msg_level, 0) +1


print(per_user)
```
