### Capturing Groups

#### Good for extracting information from a group of data like full names etc

```python
result = re.search(r"^(\w*), (\w*)$", "Lovelace, Ada")
print(result)
print(result.groups())
print(result[0])
print(result[1])
print(result[2])
"{} {}".format(result[2], result[1])
```

```python
def rearrange_name(name):
    result = re.search(r"^(\w*), (\w*)$", name)
    if result is None:
        return name
    return "{} {}".format(result[2], result[1])
rearrange_name("Lovelace, Ada")
```

```python
def rearrange_name(name):
    result = re.search(r"^(\w*), (\w*)$", name)
    if result is None:
        return name
    return "{} {}".format(result[2], result[1])
rearrange_name("Ritchie, Dennis")
```

```python
def rearrange_name(name):
    result = re.search(r"^([\w \.-]*), ([\w \.-]*)$", name)
    if result == None:
        return name
    return "{} {}".format(result[2], result[1])
rearrange_name("Hopper, Grace M.")
```

### More On Repitition Qualifyers

```python
# Search for words with exactly 5 letters
print(re.search(r"[a-zA-Z]{5}", "a ghost"))
print(re.search(r"[a-zA-Z]{5}", "a scary ghost appeared"))
# Find all words with exactly 5 letters
print(re.findall(r"[a-zA-Z]{5}", "a scary ghost appeared"))
# Find all words with exactly 5 letters using word boundaries. 
# Ignore words that are longer than 5 letters.
re.findall(r"\b[a-zA-Z]{5}\b", "A scary ghost appeared")
# Find words with length between 5 and 10 letters
print(re.findall(r"\w{5,10}", "I really like strawberries"))
# Find words with length of at least 5 letters
print(re.findall(r"\w{5,}", "I really like strawberries"))
# Find words with length up to 20 letters
print(re.search(r"s\w{,20}", "I really like strawberries"))
```

## Extracting PID Using RegExes 

```python
# () is used for grouping. i.e to create capture groups.
# [] is used for character classes. i.e to specify a set of characters.
log = "July 31 07:51:48 mycomputer bad_process[12345]: ERROR Performing package upgrade"
regex = r"\[(\d+)\]"
result = re.search(regex, log)
print(result[1])
```

```python
log = "July 31 07:51:48 mycomputer bad_process[12345]: ERROR Performing package upgrade"
regex = r"\[(\d+)\]:\s([A-Z]+)" # Added capture group for the error code
result = re.search(regex, log)
print(result [1], result[2])
result = re.search(regex, "A completely different string that also has numbers [34567]")
print(result)
print("-"*30, "\n Date extraction example\n" ,"-"*30)
date_match = re.search(r"^['\w\s]*[:\s\d]*", log)
print(date_match[0])

```

```python
log = "July 31 07:51:48 mycomputer bad_process[12345]: ERROR Performing package upgrade"
regex = r"\[(\d+)\]"
result = re.search(regex, log)
result = re.search(regex, "A completely different string that also has numbers [34567]")
result = re.search(regex, "99 elephants in a [cage]")
print(result[1])
#Note that this print command results in an error as shown in the video. 
```

```python
log = "July 31 07:51:48 mycomputer bad_process[12345]: ERROR Performing package upgrade"
regex = r"\[(\d+)\]"
result = re.search(regex, log)
result = re.search(regex, "A completely different string that also has numbers [34567]")
result = re.search(regex, "99 elephants in a [cage]")
# Check if result is None before trying to access result[1]
def extract_pid(log_line):
    regex = r"\[(\d+)\]"
    result = re.search(regex, log_line)
    if result is None:
        return "____"
    return result[1]
print(extract_pid(log))
print(extract_pid("99 elephants in a [cage]"))
```

### Splitting And Replacing

```python
# Example of re.split(). Whatever in the square brackets is used as 
# the delimiter.
re.split(r"[.?!]", "One sentence. Another one? And the last one!")
```

```python
# Because parenthesis is added, the delimiters are also 
# included in the result.
re.split(r"([.?!])", "One sentence. Another one? And the last one!")
```

```python

# re.sub substitutes all email addresses with "[REDACTED]"
re.sub(r"[\w.%+-]+@[\w.-]+", "[REDACTED]", "Received an email for go_nuts95@my.example.com")
```

```python
# The re.sub substitutes names in "Last, First" format to 
# "First Last" format. The \2 and \1 refer to the capture groups
# in the regular expression.
re.sub(r"^([\w .-]*), ([\w .-]*)$", r"\2 \1", "Lovelace, Ada")
# prints "Ada Lovelace"
```

```python
re.split(r"the|a", "One sentence. Another one? And the last one!")
```

```python
# using the exact word in the delimiter, \b must be used. 
# Meaning word boundary.
re.split(r"\bthe\b|\ba\b", "One sentence. Another one? And the last one!")
```
