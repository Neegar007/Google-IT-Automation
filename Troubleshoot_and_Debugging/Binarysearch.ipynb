## Linear And Binary Search

* Linear Search is when we search through a list one by one to find the item
we are looking for. This is the most common way. However, when we have a very 
long list, 1000 for example, our program can take longer. Hence, we use :

* Binary Search. Here, we first sort the list, then compare the name we are 
looking for with the element in the middle of the list and check if it's smaller or bigger or equal. If smaller, what we are looking for must be at the first half. If bigger, it must be at the second half.
** We take the half of the element that most like have our item, we half it again and continue until we find what we are looking for.
BINARY SERACH can bu used in so many areas like server or app troubleshooting etc, code etc

```python
# Linear Search

def linear_search(list, key):
    """If key is in the list returns its position in the list,
       otherwise returns -1."""
    for i, item in enumerate(list):
        if i == key:
            return item
    return -1

lst = [1,2,"Ade", "True",4]
key = 2
var = linear_search(lst, key)
print(var)
```

```python
# Binary Search

def binary_search(list, key):
    """Returns the position of key in the list if found, -1 otherwise.

    List must be sorted.
    """
    left = 0
    right = len(list) - 1
    while left <= right:
        middle = (left + right) // 2
        
        if list[middle] == key:
            return middle
        if list[middle] > key:
            right = middle - 1
        if list[middle] < key:
            left = middle + 1
    return -1
```
